```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Plotting the analog signal
Time = np.linspace(0, 2,100) #for a smooth plot
analogSignal = 1.4 * np.sin(2 * np.pi * 0.25 * Time)
plt.plot(Time, analogSignal)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Analog Signal')
plt.grid(True)
plt.show()
```

```python
# 3-bit encoding with time sample of 0.25s (Freq. = 4Hz)
# Defining the Time and Digital signal arrays
Time = [0, 0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2]
threeBitFourHz = [0,0.4,1,1.2,1.4, 1.2, 1, 0.4, 0]
#Plotting
plt.figure(figsize=(10, 5))
plt.step(Time, threeBitFourHz, where='post', label='Digital Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Digital Signal (3-bit, 0.25s Sampling)')
plt.grid(True)
plt.show()
```

```python
# 3-bit encoding with time sample of 0.5s (Freq. = 2Hz)
threeBitTwoHz = [0,0,1,1,1.2,1.2,1,1,0]
#Plotting
plt.figure(figsize=(10, 5))
plt.step(Time, threeBitTwoHz, where='post', label='Digital Signal')  # Use plt.step for step plot
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Digital Signal (3-bit, 0.5s Sampling)')
plt.grid(True)
plt.show()
```

```python
# 3-bit encoding with time sample of 1s (Freq. = 1Hz)
threeBitOneHz = [0,0,0,0,1,1,1,1,0]
#Plotting
plt.figure(figsize=(10, 5))
plt.step(Time, threeBitOneHz, where='post', label='Digital Signal')  # Use plt.step for step plot
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Digital Signal (3-bit, 1s Sampling)')
plt.grid(True)
plt.show()
```

```python
# 2-bit encoding with time sample of 0.25s (Freq. = 4Hz)
twoBitFourHz = [0,0.4,0.8,1.2,1.2,1.2,0.8,0.4,0]
#Plotting
plt.figure(figsize=(10, 5))
plt.step(Time, twoBitFourHz, where='post', label='Digital Signal')  # Use plt.step for step plot
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Digital Signal (2-bit, 0.25s Sampling)')
plt.grid(True)
plt.show()
```
