```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
```

```python
customers = pd.read_csv('https://raw.githubusercontent.com/BMW-M5-k/Datasets/main/Mall_Customers.csv')
customers.dropna(inplace=True)
customers = customers.drop('Gender', axis=1)
#ŸêApplying K-means clustering
cluster = KMeans(n_clusters = 5)
cluster.fit(customers)
customers['Group by KMeans'] = cluster.labels_
customers.head(20)

```

1. What type of machine learning problem is this?
- This is an unsupervised machine learning problem (clustering)

2. What machine learning algorithms are suitable for customer
segmentation tasks?
- Clustering algorithms, such as K-Means or DBSCAN

3. Plot the Answer :

```python
#Each color represents a different group of customers
plt.scatter(customers['Annual Income (k$)'], customers['Spending Score (1-100)'], c = customers['Group by KMeans'])
plt.xlabel('Annual Income ($)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
features = customers.drop(['Group by KMeans', 'CustomerID'], axis=1)
scaler = StandardScaler()
features = scaler.fit_transform(features)
search_radius = 0.6
samples = 5
dbscan = DBSCAN(eps=search_radius, min_samples=samples)
dbscan.fit(features)
customers['Group by DBSCAN'] = dbscan.labels_
```

```python
plt.scatter(customers['Annual Income (k$)'], customers['Spending Score (1-100)'], c = customers['Group by DBSCAN'])
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

- **The plots show that KMeans was more accurate at segmenting the customers than DBSCAN.**
